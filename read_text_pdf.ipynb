```python
import re

def extract_resume_sections(text):
    sections = {
        "name": None,
        "contact_links": [],
        "skills": [],
        "experience": []
    }

    lines = text.strip().split('\n')
    for i, line in enumerate(lines):
        line = line.strip()

        # NAME: Assume it's the first non-empty line
        if not sections["name"] and len(line.split()) <= 4 and re.match(r'^[A-Z ]+$', line):
            sections["name"] = line.title()

        # CONTACTS + LINKS
        if re.search(r'\b(?:email|phone|linkedin|github|portfolio|contact)\b', line.lower()):
            sections["contact_links"].append(line)

        # SKILLS
        elif 'skills' in line.lower():
            for j in range(i+1, min(i+10, len(lines))):  # Read next few lines
                if lines[j].strip() == '' or 'experience' in lines[j].lower():
                    break
                sections["skills"].append(lines[j].strip())

        # EXPERIENCE
        elif 'experience' in line.lower():
            for j in range(i+1, len(lines)):
                if lines[j].strip().lower() in ['education', 'projects', 'certifications']:
                    break
                sections["experience"].append(lines[j].strip())

    return sections


with open(r"doc\sample.txt", "r", encoding="utf-8") as f:
    text = f.read()

parsed = extract_resume_sections(text)

for section, content in parsed.items():
    print(f"\n===== {section.upper()} =====")
    if isinstance(content, list):
        for item in content:
            print(item)
    else:
        print(content)

```

```python
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Read resume text from file
with open(r'doc\sample.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Apply NLP processing
doc = nlp(text)

# Print detected named entities
for ent in doc.ents:
    print(f"{ent.text} -> {ent.label_}")
```
